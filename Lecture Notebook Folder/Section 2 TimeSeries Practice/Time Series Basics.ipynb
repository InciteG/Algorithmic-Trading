# Downsample with Resampling

```python
temp.resample('D').first().head(5)
```

```python
temp.resample('W').mean().head(5)
```

```python
temp.resample('W-Wed').mean().head(5)
```

```python
temp.resample('M').mean().head(5)
```

```python
#month by first date
temp.resample('MS').mean().head(5)
```

```python
#month by offset date
temp.resample('MS',loffset='14D').mean().head(5)
```

```python
#quarter
temp.resample('Q').mean().head(5)
```

```python
#quarter diff start date
temp.resample('Q-Feb').mean().head(5)
```

```python
#by year
temp.resample('YS').mean().head(5)
```
